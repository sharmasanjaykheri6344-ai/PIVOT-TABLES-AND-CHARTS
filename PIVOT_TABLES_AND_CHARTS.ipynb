Q1.What is the purpose of a PivotTable in Excel?

...> A PivotTable is an Excel tool that automatically summarizes, analyzes, and presents large sets of data, allowing you to quickly spot trends and patterns. For example, instead of manually calculating sales totals for each product in a spreadsheet with thousands of rows, a PivotTable can generate a clean summary in seconds.
Core purposes of a PivotTable
Summarize data: A PivotTable takes a large list of raw data and condenses it into a meaningful summary. It automatically groups and calculates totals, averages, or counts for you, eliminating the need for complex formulas.
Analyze trends: It allows you to reorganize data from different perspectives to see patterns you wouldn't notice in the raw data. By "pivoting" or rotating rows and columns, you can quickly answer business questions like "Which sales region was most profitable?" or "Which products sold best in the third quarter?".
Filter and sort interactively: You can easily filter the data to focus on a specific subset, such as sales figures for a particular month or employee. The interactive nature of a PivotTable allows you to change what you are viewing with simple drag-and-drop actions.
Create dynamic reports: Once you create a PivotTable, it serves as a dynamic report. If you add new data to your source table, you can simply "refresh" the PivotTable to update the summary instantly. This saves you from having to manually redo your analysis each time your data changes.
Identify unique values: You can use a PivotTable to create a unique list of all the values in a column, which is helpful for cleaning data and finding spelling errors.
